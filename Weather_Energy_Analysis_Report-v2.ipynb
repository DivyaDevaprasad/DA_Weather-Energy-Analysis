### 2. The maximum and minimum “load actual” of each month and each year.


```python
energy["time"] = energy["time"].apply(lambda x : pd.to_datetime(x[:19] , format= "%Y-%m-%d %H:%M:%S" ) )
```

```python
min_price = energy.groupby([energy['time'].dt.strftime('%Y'), energy['time'].dt.strftime('%m')])['price actual'].min()
max_price = energy.groupby([energy['time'].dt.strftime('%Y'), energy['time'].dt.strftime('%m')])['price actual'].max()
min_load = energy.groupby([energy['time'].dt.strftime('%Y'), energy['time'].dt.strftime('%m')])['total load actual'].min()
max_load = energy.groupby([energy['time'].dt.strftime('%Y'), energy['time'].dt.strftime('%m')])['total load actual'].max()

price_load = pd.DataFrame([min_price, max_price, min_load, max_load]).T
price_load.columns = ["Min Price actual", "Max Price actual", "Min Load actual", "Max Load actual"]

price_load
```

### 3. The “temp max” , “temp min”, “weather_main” on the time of maximum and minimum “load actual”

```python
weather["dt_iso"] = weather["dt_iso"].apply(lambda x : pd.to_datetime(x[:19] , format= "%Y-%m-%d %H:%M:%S" ) )
```

```python
weather_energy = pd.concat([weather, energy], axis = 1)

weather_desc = pd.merge( weather_energy , min_load , on = 'total load actual')
weather_desc[['total load actual', "temp_max", "temp_min", "weather_main"]]
```

```python
weather_energy = pd.concat([weather, energy], axis = 1)

weather_desc = pd.merge( weather_energy , max_load , on = 'total load actual')
weather_desc[['total load actual', "temp_max", "temp_min", "weather_main"]]
```

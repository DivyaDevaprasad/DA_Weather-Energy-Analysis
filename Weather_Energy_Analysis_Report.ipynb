## 1. The maximum and minimum “price actual” of each day and each year.

```python
energy["time"] = energy["time"].apply(lambda x : pd.to_datetime(x[:19] , format= "%Y-%m-%d %H:%M:%S" ) )
```

```python
max_price = pd.DataFrame(energy.groupby([energy['time'].dt.strftime('%Y'), energy['time'].dt.strftime('%d')])['price actual'].max())

max_price
```

```python
min_price = pd.DataFrame(energy.groupby([energy['time'].dt.strftime('%Y'), energy['time'].dt.strftime('%d')])['price actual'].min())

min_price
```

## 2. The maximum and minimum “load actual” of each day and each year

```python
max_load = pd.DataFrame(energy.groupby([energy['time'].dt.strftime('%Y'), energy['time'].dt.strftime('%d')])['total load actual'].max())

max_load
```

```python
min_load = pd.DataFrame(energy.groupby([energy['time'].dt.strftime('%Y'), energy['time'].dt.strftime('%d')])['total load actual'].min())

min_load
```

## 3. The “temp max” and “weather description” during maximum and minimum “load actual”

```python
weather["dt_iso"] = weather["dt_iso"].apply(lambda x : pd.to_datetime(x[:19] , format= "%Y-%m-%d %H:%M:%S" ) )

weather_energy = pd.concat([weather, energy], axis = 1)

weather_energy.shape
```

```python
# weather_energy[ weather_energy["total load actual"] == min_load]["temp_max"]["weather_description"]

weather_desc = pd.merge( weather_energy , min_load , on = 'total load actual')
weather_desc[['total load actual', "temp_max", "weather_description"]]
```

```python
weather_desc = pd.merge( weather_energy , max_load , on = 'total load actual')
weather_desc[['total load actual', "temp_max", "weather_description"]]
```

##  4. Maximum and minimum “generation of wind on shore” of each month, each year and “wind speed” and “wind degree” during that time

```python
min_wind = pd.DataFrame(energy.groupby([energy['time'].dt.strftime('%Y'), energy['time'].dt.strftime('%d')])['generation wind onshore'].min())

gen_min_wind = pd.merge( weather_energy , min_wind , on = 'generation wind onshore')
gen_min_wind[['generation wind onshore', "wind_speed", "wind_deg"]]
```

```python
max_wind = pd.DataFrame(energy.groupby([energy['time'].dt.strftime('%Y'), energy['time'].dt.strftime('%d')])['generation wind onshore'].max())

gen_max_wind = pd.merge( weather_energy , max_wind , on = 'generation wind onshore')
gen_max_wind[['generation wind onshore', "wind_speed", "wind_deg"]]
```
